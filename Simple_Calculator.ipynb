# Calculator Project

## Overview
This Jupyter Notebook contains the implementation of a simple calculator. The calculator can perform basic arithmetic operations such as addition, subtraction, multiplication, and division.

## Features
- Addition of two numbers
- Subtraction of two numbers
- Multiplication of two numbers
- Division of two numbers
- Exit the calculator

## Usage
To use the calculator, simply run the cells in the notebook. Each cell contains a function for a specific arithmetic operation. You can modify the input values to perform different calculations.

## Requirements
This project requires Python 3.0 or plus and Jupyter Notebook to run.

```python
def Simple_Calculator():
    while True:
        print("Welcome to the Simple Calculator \n------------------------------ \nPlease select the operation you want to perform \n------------------------------ \n1. Addition \n------------------------------ \n2. Subtraction \n------------------------------ \n3. Multiplication \n------------------------------ \n4. Division \n------------------------------ \n5. Exit \n------------------------------")

        operation = input("Enter the operation you want to perform (1, 2, 3, 4, 5): ")

        if operation == '5':
            print("Exiting the calculator. Goodbye!")
            break

        if operation in ('1', '2', '3', '4'):
            try:
                num1 = float(input("Enter the first number: "))
                num2 = float(input("Enter the second number: "))
            except ValueError:
                print("Invalid input. \nPlease enter a valid number")
                continue
            
            if operation == '1':
                print("Operation Selected: Addition \n------------------------------")
                print(f"Solution :- {num1} + {num2} = {num1 + num2} \n------------------------------")
            elif operation == '2':
                print("Operation Selected: Subtraction \n------------------------------")
                print(f"Solution :- {num1} - {num2} = {num1 - num2} \n------------------------------")
            elif operation == '3':
                print("Operation Selected: Multiplication \n------------------------------")
                print(f"Solution :- {num1} x {num2} = {num1 * num2} \n------------------------------")
            elif operation == '4':
                try:
                    print("Operation Selected: Division \n------------------------------")
                    print(f"Solution :- {num1} รท {num2} = {num1 / num2} \n------------------------------")
                except ZeroDivisionError:
                    print("Cannot divide by zero")

        else:
            print("Invalid input. \nPlease enter a valid operation")

if __name__ == "__main__":
    Simple_Calculator()                    
```
